```python
import pandas as pd
```

```python
afm_dataFrame = pd.read_csv("AFM_data.csv")
rmc_dataFrame = pd.read_csv("RMC_data.csv", low_memory=False)
```

```python
print("Unique Product-Id in AFM data - {}".format(len(afm_dataFrame['product_id'].unique())))
print("Unique Product-Id in RMC data - {}".format(len(rmc_dataFrame['product_id'].unique())))
```

```python
common_columns = rmc_dataFrame.columns.intersection(afm_dataFrame.columns)
print("Columns intersecting in both dataset - {}".format(common_columns))
```

```python
merged_dataFrame = pd.merge(afm_dataFrame, rmc_dataFrame, on = 'product_id')
merged_dataFrame.head()
```

```python
merged_dataFrame.shape
```

```python
unique_values_dict = dict()
for column in merged_dataFrame.columns:
    print("{} - {}".format(column, merged_dataFrame[column].unique().tolist()))
    unique_values_dict[column] = merged_dataFrame[column].unique().tolist()
    print()
```

```python
import numpy as np
dataFrame_copy = merged_dataFrame.copy()
unknown_values = ["NULL", "null", "Null", "unknown", "Unknown", "Nan", "nan", "NaN"]
for key, value in unique_values_dict.items():
    if len(value) == 1 and value[0] in unknown_values:
        dataFrame_copy[key] = dataFrame_copy[key].replace(value[0], None).astype(object)
        print("Changed datatype of a column - {}".format(key))
    else:
        for word in value:
            if word in unknown_values:
                dataFrame_copy[key] = dataFrame_copy[key].replace(word, "").astype(object)
                print("Changed datatype of a value in column - {}".format(key))
                break
```

```python
dataFrame_copy.info()
```

```python
preprocessed_dataFrame = dataFrame_copy.dropna(axis = 1)
preprocessed_dataFrame.head()
```

```python
preprocessed_dataFrame.shape
```

```python
preprocessed_dataFrame.info()
```

```python
unique_product_ids = preprocessed_dataFrame['product_id'].unique()
print("Unique Product-Id : ", unique_product_ids)
```

```python
preprocessed_dataFrame.to_csv("Preprocessed Dataset.csv", index=False)
print("Saved Preprocessed dataframe")
```
